```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Define the URL for iPhone listings (Flipkart example)
url = "https://www.flipkart.com/search?q=iphone"

# Set headers to mimic a real browser
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}
```

```python
# Send a GET request
response = requests.get(url, headers=headers)
```

```python
# Check for successful request
if response.status_code == 200:
    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Print and inspect the HTML structure
    print(soup.prettify()[:1000])  # Print first 1000 characters for inspection
    
    # Extract product details
    products = []
    for item in soup.find_all("div", class_="_1AtVbE"):
        name = item.find("div", class_="_4rR01T")
        price = item.find("div", class_="_30jeq3")
        rating = item.find("div", class_="_3LWZlK")
        reviews = item.find("span", class_="_2_R_DZ")
        availability = "Available" if name else "Out of Stock"
        
        if name and price:
            products.append({
                "Product Name": name.text,
                "Price": price.text,
                "Rating": rating.text if rating else "N/A",
                "Reviews": reviews.text if reviews else "N/A",
                "Availability": availability
            })
    
    # Convert to DataFrame
    df = pd.DataFrame(products)
    
    # Save to CSV
    df.to_csv("iphone_listings.csv", index=False)
    print("Data saved to iphone_listings.csv")
else:
    print(f"Failed to fetch data, status code: {response.status_code}")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

# Function to scrape iPhone data
def scrape_iphones(url, pages=3):
    iphone_data = []

    for page in range(1, pages + 1):
        print(f"Scraping page {page}...")
        response = requests.get(url.format(page))
        if response.status_code != 200:
            print(f"Failed to fetch page {page}")
            continue

        soup = BeautifulSoup(response.text, 'html.parser')

        products = soup.find_all('div', class_='some-product-class')  # Update this based on website structure

        for product in products:
            try:
                name = product.find('a', class_='product-name-class').text.strip()
            except AttributeError:
                name = "N/A"

            try:
                price = product.find('div', class_='product-price-class').text.strip()
            except AttributeError:
                price = "N/A"

            try:
                ratings = product.find('span', class_='ratings-class').text.strip()
            except AttributeError:
                ratings = "N/A"

            try:
                reviews = product.find('span', class_='reviews-class').text.strip()
            except AttributeError:
                reviews = "N/A"

            try:
                availability = product.find('span', class_='availability-class').text.strip()
            except AttributeError:
                availability = "N/A"

            iphone_data.append({
                "Product Name": name,
                "Price": price,
                "Ratings": ratings,
                "Number of Reviews": reviews,
                "Availability": availability
            })

        time.sleep(2)  # Avoid getting blocked

    return iphone_data

# Flipkart or BestBuy URL (Modify as needed)
base_url = "https://www.example.com/search?q=iphone&page={}"  # Replace with actual URL

# Scrape iPhone listings
iphone_list = scrape_iphones(base_url)

# Convert to DataFrame
df = pd.DataFrame(iphone_list)

# Save as CSV
df.to_csv("iphones_data.csv", index=False)
print("\nData scraping complete. Saved as iphones_data.csv.")

# Exploratory Data Analysis (EDA)
print("\nBasic Exploratory Data Analysis:")
print("Total iPhones scraped:", len(df))

# Verify available columns
print("\nAvailable columns:", df.columns)

# Price Distribution - If available
if "Price" in df.columns and df["Price"].notna().sum() > 0:
    print("\nPrice Distribution:")
    print(df["Price"].value_counts())
else:
    print("\nPrice data is missing. Showing Ratings Distribution instead:")
    if "Ratings" in df.columns and df["Ratings"].notna().sum() > 0:
        print(df["Ratings"].value_counts())
    else:
        print("Ratings data is also missing.")

# Average Rating - If available
if "Ratings" in df.columns and df["Ratings"].notna().sum() > 0:
    try:
        df["Ratings"] = df["Ratings"].astype(float)
        print("\nAverage Rating:", df["Ratings"].mean())
    except ValueError:
        print("\nRatings column contains non-numeric values.")

# Most Reviewed Products - If available
if "Number of Reviews" in df.columns and df["Number of Reviews"].notna().sum() > 0:
    print("\nTop Reviewed Products:")
    print(df.nlargest(5, "Number of Reviews")[["Product Name", "Number of Reviews"]])
else:
    print("\nNumber of Reviews data is missing.")
```
